```python
import yfinance as yf
import pandas as pd
appleStock = yf.Ticker("AAPL")
appleStock = appleStock.history(start='2018-01-01', end='2019-12-9')
appleOpenPrices = appleStock['Open'].tolist()
appleDates = appleStock.index.tolist()
microsoftStock = yf.Ticker("MSFT")
microsoftStock = microsoftStock.history(start='2018-01-01', end='2019-12-9')
microsoftOpenPrices = microsoftStock['Open'].tolist()
microsoftDates = microsoftStock.index.tolist()
teslaStock = yf.Ticker("TSLA")
teslaStock = teslaStock.history(start='2018-01-01', end='2019-12-9')
teslaOpenPrices = teslaStock['Open'].tolist()
teslaDates = teslaStock.index.tolist()
```

```python
frame={
    'date':appleDates,
    'apple':appleOpenPrices,
    'microsoft':microsoftOpenPrices,
    'tesla':teslaOpenPrices
}
df=pd.DataFrame(frame,index=appleDates)
df
```

```python
import plotly.express as px
fig = px.line(df, x='date', y=['apple','microsoft','tesla'])
fig.show()
```
